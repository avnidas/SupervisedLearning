```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from scipy.stats import zscore
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("abalone.csv",header=None,names=['Gender','Length','Diameter','Height','Weight','MeatWeight','GutWeight','ShellWeight','Rings'])
```

```python
df.head()
```

```python
df['Gender'].value_counts()
```

```python
df_rows , df_cols = df.shape
print(df_rows)
print(df_cols)
```

```python
df2 = df.loc[:, 'Length':'Rings']
```

```python
df2.head()
```

```python
import seaborn as sns
sns.pairplot(df2,diag_kind='kde')
```

```python
df_scaled = df2.apply(zscore)
```

```python
df_scaled.head()
```

```python
model = KMeans(n_clusters = 3)
```

```python
model
```

```python
cluster_range = range( 1, 15 )
cluster_errors = []
for num_clusters in cluster_range:
  clusters = KMeans( num_clusters, n_init = 10 )
  clusters.fit(df_scaled)
  labels = clusters.labels_
  centroids = clusters.cluster_centers_
  cluster_errors.append( clusters.inertia_ )
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
clusters_df[0:15]
```

The total sum of squared distances of every data point from respective centroid is also called inertia. Let us print the inertia value for all K values. That K at which the inertia stop to drop significantly (elbow method) will be the best K.

```python
# Elbow plot

plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
```

```python
from scipy.spatial.distance import pdist
```

```python
df_scaled['labels'].value
```

```python
# cophenet index is a measure of the correlation between the distance of points in feature space and distance on dendrogram
# closer it is to 1, the better is the clustering

Z = linkage(df_scaled, 'average')#Average is having good score than ward
c, coph_dists = cophenet(Z , pdist(df_scaled))

c
```

```python
plt.figure(figsize=(20, 10))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90., color_threshold=3, leaf_font_size=10. )
plt.tight_layout()
```
